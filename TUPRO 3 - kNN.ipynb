# Rumus Perhitungan jarak dengan Euclidean, Mahattan, Minkowski, dan Supremum

```python
# Fungsi untuk menghitung jarak dengan Euclidean
def euclidean_distance(dataTrain, dataTest):
    dist = 0
    for i in range(len(dataTest)):
        dist += (dataTrain[i] - dataTest[i])**2

    return dist ** (1/2)

# Fungsi untuk menghitung jarak dengan Manhattan
def manhattan_distance(dataTrain, dataTest):
    dist = 0
    for i in range(len(dataTrain)):
        dist += abs(dataTrain[i] - dataTest[i])

    return dist
 
# Fungsi untuk menghitung jarak dengan Minkowski
def minkowski_distance(dataTrain, dataTest, p=2):
    dist = 0
    for i in range(len(dataTrain)):
        dist += abs(dataTrain[i] - dataTest[i]) ** p

    return dist ** (1/p)

# Fungsi untuk menghitung jarak dengan Supremum
def supremum_distance(dataTrain, dataTest):
    dist = []
    for i in range(len(dataTrain)):
        dist.append(abs(dataTrain[i] - dataTest[i]))

    return max(dist)
```

# Proses Mendapatkan Jarak dari masing-masing Metric Distance

```python
# Fungsi untuk mendapatkan nilai jarak dari masing-masing Metric Distance

def getDistances(dfAtribut, dfNama, dfTest):
    
    k = 3

    euclidean, manhattan, minkowski, supremum = [],[],[],[]

    for idx, row in dfAtribut.iterrows():
        euclidean.append([euclidean_distance(row, dfTest), dfNama[idx]])
        manhattan.append([manhattan_distance(row, dfTest), dfNama[idx]])
        minkowski.append([minkowski_distance(row, dfTest), dfNama[idx]])
        supremum.append([supremum_distance(row, dfTest), dfNama[idx]])

    
    euclidean   = sorted(euclidean)[:k]
    manhattan   = sorted(manhattan)[:k]
    minkowski   = sorted(minkowski)[:k]
    supremum    = sorted(supremum)[:k]


    return euclidean, manhattan, minkowski, supremum
```

```python
# Menyediakan List kosong untuk menampung masing-masing jarak dari setiap Metric Distance
euclidean, manhattan, minkowski, supremum = [],[],[],[]

# Proses Inputan Data Test berupa : ukuran, kenyamanan, irit, kecepatan, harga
print("Masukan Data Testing : ")
ukuran = int(input('Ukuran [1 - 10] = '))
kenyamanan = int(input('Kenyamanan [1 - 10] = '))
irit = int(input('Irit [1 - 10] = '))
kecepatan= int(input('Kecepatan [1 - 10] = '))
harga = float(input('Harga (Ratus Juta) [1.0 - 10.0] = '))

# Data Test diatas disimpan pada List dataTest
dataTest = [ukuran, kenyamanan, irit, kecepatan, harga]

# Proses Normalisasi atribut harga pada dataTest
dataTest = NormalisasiforTest(dataTest, dfTemp)

# Proses menyimpan jarak dari masing-masing Metric Distance mengguanakan Prosedur getDistances
euclidean, manhattan, minkowski, supremum = getDistances(dfAtribut, dfNama, dataTest)

# Masing-masing Metric Distance disimpan pada Disctionary jarak
jarak = {'Euclidean' : euclidean, 'Manhattan':manhattan, 'Minkowski':minkowski, 'Supremum':supremum}

# Menyediakan variabel rekomendasi untuk ditambahkan kedalam File Excel 'rekomendasi.xls'
rekomendasi = {}

# Proses menampilkan 3 Nama Mobil Rekomendasi dari masing-masing Metric Distance beserta jaraknya
for metric, items in jarak.items():
    nama = []
    print(f'\nRekomendasi dengan {metric} Distance')
    print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')
    for atr in items:
        print(f'Mobil {atr[1]} dengan Distance : {round(atr[0], 2)}')
        nama.append(atr[1])
        
    rekomendasi[metric] = nama

# Memasukan Data Frame dari variabel 'rekomendasi' kedalam File Excel 'rekomendasi.xls'
dfRekomendasi = pd.DataFrame(rekomendasi)
dfRekomendasi.to_excel("rekomendasi.xls")

```
